```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
t
```

```python
s = 1 + np.sin(2 * np.pi * t)
s
```

```python
fig, ax = plt.subplots()
ax.plot(t, s)
```

```python
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()
```

```python
fig.savefig("test.png")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```
